```python
import pandas as pd
import numpy as np
from arch import arch_model
```

```python
df_data = pd.read_csv('Data.csv', header=0, sep=';', index_col=0, parse_dates=True)
```

```python
date_columns = df_data.columns[0::2]
price_columns = df_data.columns[1::2]

# Créer un dictionnaire pour stocker les DataFrames individuels
dataframes = {}

for date_col, price_col in zip(date_columns, price_columns):
    temp_df = df_data[[date_col, price_col]].dropna()
    temp_df.columns = ['Date', price_col]
    temp_df['Date'] = pd.to_datetime(temp_df['Date'], format='%d/%m/%Y', errors='coerce')
    dataframes[price_col] = temp_df

# Fusionner tous les DataFrames sur la colonne 'Date'
merged_df = pd.DataFrame({'Date': pd.to_datetime([])})

for price_col, temp_df in dataframes.items():
    merged_df = pd.merge(merged_df, temp_df, on='Date', how='outer')

# Trier par date
merged_df = merged_df.sort_values(by='Date').reset_index(drop=True)

# Supprimer les colonnes Unnamed: 27 et Unnamed: 29
merged_df = merged_df.drop(['Unnamed: 27', 'Unnamed: 29'], axis=1)

# Extract csv file merged
merged_df.to_csv('merged.csv', index=False)
```

```python

# Example of using a GARCH model to predict volatility
# Generate some random data
np.random.seed(42)
returns = np.random.normal(0, 1, 1000)

# Fit the GARCH(1,1) model
model = arch_model(returns, vol='Garch', p=1, q=1)
model_fit = model.fit()

# Make volatility prediction
forecast = model_fit.forecast(horizon=5)
print(forecast.variance[-1:])
# Calculer la volatilité pour chaque actif
volatility_results = {}

for asset, df in dataframes.items():
    returns = df[asset].pct_change().dropna()
    model = arch_model(returns, vol='Garch', p=1, q=1)
    model_fit = model.fit(disp='off')
    forecast = model_fit.forecast(horizon=5)
    volatility_results[asset] = forecast.variance.iloc[-1].values

# Convertir les résultats en DataFrame
volatility_df = pd.DataFrame(volatility_results, index=[f'T+{i+1}' for i in range(5)])
print(volatility_df)
```
